# Density Matrices

Consider a quantum system composed by $N$ subsystems (spins, atoms, particles, etc...) each described by a wave function $\psi_i \in \mathcal{H}^D$, where $\mathcal{H}^D$ is a $D$-dimensional Hilbert space. How do you write the total wave function of the system $\Psi(\psi_1, \psi_2, \dots, \psi_N)$?

1. Write a code (Fortran or Python) to describe the composite system in the case of $N$-body non-interacting, separable pure state;

2. and in the case of a general $N$-body pure wave function $\Psi \in \mathcal{H}^{DN}$;

3. Comment and compare their efficiency;

4. Given $N=2$, write the density matrix of a general pure state $\Psi$, $\rho = |\Psi\rangle\langle\Psi|$;

5. Given a generic density matrix of dimension $D^N \times D^N$, compute the reduced density matrix of either the left or the right system, e.g., $\rho_1 = \text{Tr}_2 \rho$.

6. Test the functions described before (and all others needed) on two-spin one-half (qubits) with different states.

```python
# Set dimensions
N = 2
D = 2

# Define basis states
zero = np.array([1, 0])
one = np.array([0, 1])
plus = zero + one
minus = zero - one

# Build separable states
sep_1 = dm.separable_state(N, D, [zero, plus])        # |0>|+>
sep_2 = dm.separable_state(N, D, [one, minus])        # |1>|->
sep_3 = dm.separable_state(N, D, [plus, plus])        # |+>|+>
sep_4 = dm.separable_state(N, D, [minus, minus])      # |->|->

# Build general states (entangled)
gen_1 = dm.general_state(N, D, (np.kron(zero, zero) + np.kron(one, one)))   # Bell state
gen_2 = dm.general_state(N, D, (np.kron(zero, one) + np.kron(one, zero)))   # Bell state
gen_3 = dm.general_state(N, D, (np.kron(zero, zero) + np.kron(one, one)))   # GHZ state
gen_4 = dm.general_state(N, D, (np.kron(zero, zero) + np.kron(one, zero) + np.kron(one, one))) # Asymmetric state
```

```python
dm.test_state(sep_1, N, D, expected_rdm_0=np.array([[1, 0], [0, 0]]), 
                           expected_rdm_1=np.array([[0.5, 0.5], [0.5, 0.5]]))

dm.test_state(sep_2, N, D, expected_rdm_0=np.array([[0, 0], [0, 1]]), 
                           expected_rdm_1=np.array([[0.5, -0.5], [-0.5, 0.5]]))

dm.test_state(sep_3, N, D, expected_rdm_0=np.array([[0.5, 0.5], [0.5, 0.5]]), 
                           expected_rdm_1=np.array([[0.5, 0.5], [0.5, 0.5]]))

dm.test_state(sep_4, N, D, expected_rdm_0=np.array([[0.5, -0.5], [-0.5, 0.5]]), 
                           expected_rdm_1=np.array([[0.5, -0.5], [-0.5, 0.5]]))
```

```python
dm.test_state(gen_1, N, D, expected_rdm_0=np.array([[0.5, 0], [0, 0.5]]),
                           expected_rdm_1=np.array([[0.5, 0], [0, 0.5]]))

dm.test_state(gen_2, N, D, expected_rdm_0=np.array([[0.5, 0], [0, 0.5]]),
                           expected_rdm_1=np.array([[0.5, 0], [0, 0.5]]))

dm.test_state(gen_3, N, D, expected_rdm_0=np.array([[0.5, 0], [0, 0.5]]),
                           expected_rdm_1=np.array([[0.5, 0], [0, 0.5]]))

dm.test_state(gen_4, N, D, expected_rdm_0=np.array([[1/3, 1/3], [1/3, 2/3]]),
                           expected_rdm_1=np.array([[2/3, 1/3], [1/3, 1/3]]))
```
